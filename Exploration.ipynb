## Pipeline probas

```python
def cb_fit_casitos_withdomprobas(self, get_gen):
    pi_no_views = train_bought(filter(lambda x : view_len(x)==0, ds(TRAIN_SPLIT_DATA)), lambda x : x)
    pd_no_views = train_bought(filter(lambda x : view_len(x)==0, ds(TRAIN_SPLIT_DATA)), dom)
    self.i_noviews = list(zip(*itertools.islice(calc_iv(pd_no_views, pi_no_views),10)))[0]

    self.boughts = defaultdict(Counter)
    for x in get_gen():
        i = x['item_bought']
        self.boughts[dom(i)][i]+=1
    self.boughts_totals = { d: sum(itemcounter.values()) for d,itemcounter in self.boughts.items() }

    pi_no_dom = train_bought(filter(lambda x : not is_viewed_domain(x), get_gen()), lambda x : x)
    pd_no_dom = train_bought(filter(lambda x : not is_viewed_domain(x), get_gen()), dom)
    self.ev_nodom = OrderedDict(calc_iv(pd_no_dom, pi_no_dom))
    
def obtain_dom_probas(self):
    bid = self.get_bucket()
    #print(bid)
    line = next(self.dom_probas_iterators[bid])
    return json.loads(line)
    
def cb_proc_casitos_withdomprobas(self, item):
    self.iviews = [v for v in views(item) if dom(v['event_info']) is not None]
    iids = [v['event_info'] for v in self.iviews]
    icount = Counter(iids)
    self.dids = [dom(iid) for iid in iids]
    dcount = Counter(self.dids)
    views_len = len(self.iviews)
    if views_len>0:
        proba_domain_in_view = 0.5275938818
        self.p_dom = obtain_dom_probas(self)
        proba_item_in_view = 0.5956234027
        p_items_inview = { k: proba_item_in_view*v/dcount[dom(k)] for k,v in icount.items() }
        self.p_items = {}
        for d in set(self.dids):
            td = self.boughts_totals.get(d,0)
            if td>0:
                for i, cant in self.boughts[d].items():
                    self.p_items[i] = (1-proba_item_in_view)*cant/td + p_items_inview.get(i,0)
            if td<10:
                for i in domain2items[d][:10-td]:
                    self.p_items[i] = 0
        for i,pi in p_items_inview.items():
            if i not in self.p_items:
                self.p_items[i]=pi
        self.p_items=list(self.p_items.items())
```

```python
def open_buckets_trivial(mode): return [open('trivial_buckets.jl', mode) for _ in range(1)]
def get_bucket_trivial(self): return 0
```

```python
def open_buckets_bydistinctdom_trivial(mode): return [open(f'trivial_bybucket_{i}.jl', mode) for i in range(3)]
def get_bucket_bydistinctdom_trivial_aux(ld):
    if 1<=ld<=2: return 0
    if 3<=ld<=7: return 1
    return 2
def get_bucket_bydistinctdom_trivial(self):
    return get_bucket_bydistinctdom_trivial_aux(len(self.dids))
    
    
    
```

```python
files = open_buckets_bydistinctdom_trivial('w')
for x in ds(VALID_SPLIT_DATA):
    vs = [v for v in views(x) if dom(v['event_info']) is not None]
    views_len = len(vs)
    if views_len>0:
        dids = [dom(v['event_info']) for v in vs]
        dids = [d for d in dids if d is not None]
        dcount = Counter(dids)
        pk = 0.5275938818
        p_dom = { k: pk*v/views_len for k,v in dcount.items() }
        p_dom[''] = (1-pk)
        bid = get_bucket_bydistinctdom_trivial_aux(len(dids))
        files[bid].write(json.dumps(p_dom)+'\n')
for f in files:
    f.close()
```
